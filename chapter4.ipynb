```python
# coding: utf-8
import MeCab
fname = 'neko.txt'
fpname = 'neko.txt.mecab'
def parse_neko():
    with open(fname, encoding='utf-8') as data_file, open(fpname, mode='w', encoding='utf-8') as out_file:
        mecab = MeCab.Tagger()
        out_file.write(mecab.parse(data_file.read()))

def neco_lines():
    with open(fpname, encoding='utf-8') as file_parsed:
        morphemes = []
        for line in file_parsed:
            cols = line.split('\t')
            if(len(cols) < 2):
                raise StopIteration     # 区切りがなければ終了
            res_cols = cols[1].split(',')
            morpheme = {
                'surface': cols[0],
                'base': res_cols[6],
                'pos': res_cols[0],
                'pos1': res_cols[1]
            }
            morphemes.append(morpheme)
            # 品詞細分類1が'句点'なら文の終わりと判定
            if res_cols[1] == '句点':
                yield morphemes
                morphemes = []

parse_neko()
lines = neco_lines()
for line in lines:
    print(line)
```
